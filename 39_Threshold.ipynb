<font color="blue" size="6">Threshold Based Dropping</font>

Another way to configure how pandas drops missing values is the thresh argument. thresh requires a minimum number of non-null values for the row/column to be kept.

```python
import numpy as np
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample1.csv")
df = pd.DataFrame(data)
df
```

```python
df.dropna(thresh=df.shape[1]-1)
```

```python
#Drops rows in a data frame with atmost x number of missing values in the rows.
df.dropna(thresh=df.shape[1]-2) #the atmost value is =2
```
