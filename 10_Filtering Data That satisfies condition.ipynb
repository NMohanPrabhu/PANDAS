 <font color="blue" size="12"><b>Filtering Data that satisfies a condition </b><font>
    

<font color ="green" size="6">Basic Steps:</font>
<p>1.Create a Boolean mask with a condition</p>
<p>2.Pass the Boolean mask to loc method to get the desired results.</p>

```python
import pandas as pd
import numpy as np
df=pd.read_csv("DATASETS\\scores.csv") 
df
```

```python
#row Boolean mask selects player who scores 100 >25 times
century=df["100"]>25
century
```

```python
df.loc[(century)]
```

```python
#selects player with strike rate >90
strike=df["SR"]>90
```

```python
strike
```

```python
#col Boolean mask selects colums starts with P
player=df.columns.str.startswith('P')
```

```python
player
```

```python
#or condition ie player who scored 100 > 25 times or strike rate > 90
df.loc[(century|strike),:]
```

```python
#and condition ie player who scored 100 > 25 times and strike rate > 90
df.loc[(century&strike),:]
```

```python
#and condition and displayes player column values that satisfies this condition
df.loc[(century&strike),player]
```

<p>Pandas where() method is used to check a data frame for one or more conditions and return the result accordingly.</p>
<p>By default, The rows not satisfying the condition are filled with NaN value.</p>

```python
#replaces with NAN that doesn’t satisfies condition
df.where(century&strike)
```

```python
#fill nan with zeros
df.where(century&strike).fillna(0)
```

```python
#drop nan fields
df.where(century&strike).dropna()
```
