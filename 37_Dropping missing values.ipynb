<font color="green" size="4">Dropping missing values using dropna()</font><p>
In order to drop a null values from a dataframe, we used dropna() function this function drop Rows/Columns of datasets with Null values in different ways. 


<font color="yellow" size="4" >Code #1: Dropping rows with at least 1 null value. </font>

```python
# importing pandas as pd 
import pandas as pd   
# importing numpy as np 
import numpy as np   
# dictionary of lists 
dict = {'First Score':[100, 90, np.nan, 95], 
        'Second Score': [30, np.nan, 45, 56], 
        'Third Score':[52, 40, 80, 98], 
        'Fourth Score':[np.nan, np.nan, np.nan, 65]} 
  # creating a dataframe from dictionary 
df = pd.DataFrame(dict) 
df 
```

```python
# using dropna() function   
df.dropna() 
```

<font color="yellow" size="4" >Code #2: Dropping rows if all values in that row are missing. </font>

```python
# importing pandas as pd 
import pandas as pd 
  
# importing numpy as np 
import numpy as np 
  
# dictionary of lists 
dict = {'First Score':[100, np.nan, np.nan, 95], 
        'Second Score': [30, np.nan, 45, 56], 
        'Third Score':[52, np.nan, 80, 98], 
        'Fourth Score':[np.nan, np.nan, np.nan, 65]} 
  
# creating a dataframe from dictionary 
df = pd.DataFrame(dict) 
    
df 
```

```python
# using dropna() function     
df.dropna(how = 'all') 
```

<font color="yellow" size="4" >Code #3: Dropping columns with at least 1 null value.</font> 

```python
# importing pandas as pd 
import pandas as pd 
   
# importing numpy as np 
import numpy as np 
   
# dictionary of lists 
dict = {'First Score':[100, np.nan, np.nan, 95], 
        'Second Score': [30, np.nan, 45, 56], 
        'Third Score':[52, np.nan, 80, 98], 
        'Fourth Score':[60, 67, 68, 65]} 
  
# creating a dataframe from dictionary  
df = pd.DataFrame(dict) 
     
df 
```

```python
# using dropna() function      
df.dropna(axis = 1) 
```

<font color="yellow" size="4" >  Code #4: Dropping Rows with at least 1 null value in CSV file </font>

```python
# importing pandas module  
import pandas as pd  
    
# making data frame from csv file  
data = pd.read_csv("DATASETS\\employees.csv")  
    
# making new data frame with dropped NA values  
new_data = data.dropna(axis = 0, how ='any')  
    
new_data 
```
