```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("CSE_RESULT.csv")
df
```

```python
df.columns
```

```python
#Find total number of all clear and failed students
result_count = df["Result"].value_counts()
print(result_count)
```

```python
#Find percentage of all clear and failed students
result_count_percent = df["Result"].value_counts(normalize = True)*100
print(result_count_percent)
```

```python
#Determine the CGPA of all the students
df.eval("CGPA=((S1_GP*S1_C)+(S2_GP*S2_C)+(S3_GP*S3_C)+(S4_GP*S4_C)+(S5_GP*S5_C)+(S6_GP*S6_C)+(L1_GP*L1_C)+(L2_GP*L2_C))/Tot_Credit", inplace=True)
df[["Name","CGPA"]]
```

```python
#List out top 5 students CGPA
top_5 = df[["Name","CGPA"]].nlargest(5,"CGPA")
print(top_5)
```

```python
#Determine which type of students (Hostle vs Dayscholar) from a particular class (SEC A vs SEC B) have performed well ie. Pass
print(df.groupby(["Sec","Type"])["Result"].apply(lambda x:(x=="PASS").mean()*100).round(2))
```

```python
#Determine which Gender (Male /Female) of students from a particular class (SEC A vs SEC B) have performed well ie. Pass
print(df.groupby(["Sec","Gender"])["Result"].apply(lambda x:(x=="PASS").mean()*100).round(2))
```

```python
#Determine which subject (Sub_1) staff from class A and class B have produced better result.Justify your answer
print(df.groupby(["Sec","S1_Staff"])[""].apply(lambda x:(x>0).sum()))
```
