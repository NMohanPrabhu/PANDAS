```python
import numpy as np
import pandas as pd
#Load The data Frame
df=pd.read_csv("DATASETS\\scores.csv") 
df
```

```python
#displays all columns
df.columns
```

```python
#Rename the column name
#It will not show the updation when displayed

df.rename(columns={'0': 'Duck Out'})
df.columns
```

```python
#Now the updation will take place updation when displayed [inplace=True]

df.rename(columns={'0': 'Duck Out'},inplace=True)
```

```python
df.columns
```

```python
#Axis columns to upper case is updated

df.rename(str.upper,axis='columns').head()
```

```python
#replace all columns with _

df.rename(lambda x: x.replace(" ","_"),axis='columns').head()
```

```python
#replace all columns with tilte case

df.rename(lambda x: x.title(),axis='columns').head()
```
