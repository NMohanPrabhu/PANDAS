<p>transform() : which transforms every value of the DataFrame concerning the provided function. 
<p>Transform is an operation used in conjunction with groupby 

```python
import pandas as pd
import numpy as np
df = pd.read_csv('DATASETS\\titanic.csv')
```

```python
df
```

```python
df["FARE_MEAN"]=df.groupby('Pclass')["Fare"].transform('mean')
```

```python
df
```

```python
df["Fare_Class_Total"]=df.groupby('Pclass')["Fare"].transform('sum')
```

```python
df
```

```python
df["Fare_perc"]=df["Fare"]/df["Fare_Class_Total"]
```
