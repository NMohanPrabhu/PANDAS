<p>Types of Joins in Pandas
<p>We will use these two Dataframes to understand the different types of joins.

<font color ="green"><b>1.Inner Join</b></font>
<p>	Inner join is the most common type of join youâ€™ll be working with. It returns a Dataframe with only those rows that have common characteristics. This is similar to the intersection of two sets.</p>
<font color ="green"><b>2.Left Join</b></font>
<p>	With a left outer join, all the records from the first Dataframe will be displayed, 	irrespective of whether the keys in the first Dataframe can be found in the second 	Dataframe. Whereas, for the second Dataframe, only the records with the keys in the 	second Dataframe that can be found in the first Dataframe will be displayed.</p>
<font color ="green"><b>3.Right Outer Join</b></font>
<p>	For a right join, all the records from the second Dataframe will be displayed. However, 	only the records with the keys in the first Dataframe that can be found in the second 	Dataframe will be displayed.</p>
<font color ="green"><b>4.Full Outer Join</b></font>
<p>	A full outer join returns all the rows from the left Dataframe, and all the rows from the 	right Dataframe, and matches up rows where possible, with NaNs elsewhere. But if the 	Dataframe is complete, then we get the same output.</p>
<font color ="green"><b>5.Index Join</b></font>
<p>	To merge the Dataframe on indices pass the left_index and right_index arguments as 	True i.e. both the Dataframes are merged on an index using default Inner Join.</p>


```python
import pandas as pd
import numpy as np
S1=pd.read_csv('DATASETS\\Student1.csv')
S2=pd.read_csv('DATASETS\\Student2.csv')
S1 = pd.DataFrame(S1)
S2 = pd.DataFrame(S2)
```

```python
S1
```

```python
S2
```

```python
Student_Left=pd.merge(S1,S2,on='Name',how='left')
Student_Left
```

```python
Student_right=pd.merge(S1,S2,on='Name',how='right')
Student_right
```

```python
Student_inner=pd.merge(S1,S2,on='Name',how='inner')
Student_inner
```

```python
Student_outer=pd.merge(S1,S2,on='Name',how='outer')
Student_outer
```

```python
Student_index=pd.merge(S1,S2,left_index=True, right_index=True)
Student_index
```
