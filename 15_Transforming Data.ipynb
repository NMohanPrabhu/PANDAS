<font color ="green" size="5"><b> Assign() : </b> </font>
   <p> Dataframe.assign() method assign new columns to a DataFrame, returning a new object (a copy) 
    with the new columns added to the original ones. </p>
   <p> Existing columns that are re-assigned will be overwritten.</p>
   <p> Length of newly assigned column must match the number of rows in the dataframe.</p>

```python
import numpy as np
import pandas as pd
df=pd.read_csv("DATASETS\\Tips.csv") 
df
```

```python
df=df.assign(tip_per = lambda x: round(x["Tips"]/x["Bill"],2), tip_per_person = lambda x: round(x["Tips"]/x["Size"],2))
```

<font color ="green" size ="4" ><b> applymap():</b></font>
<p>The applymap() function is used to apply a function to a Dataframe elementwise.
<p>This method applies a function that accepts and returns a scalar to every element of a DataFrame.


<font color="grey" ><b>#Example_1</font>

```python
import numpy as np
import pandas as pd
#Load The data Frame
df=pd.read_csv("DATASETS\\scores.csv") 
df
```

```python
# Using lambda function we first convert all  
# the cell to a string value and then find 
# its length using len() function 
df.applymap(lambda x: len(str(x))) 
```

```python
# Using applymap() to append '_X' 
# in each cell of the dataframe 
df.applymap(lambda x: str(x) + '_X') 
```

```python
df.applymap(lambda x: str.lower(x) if str(x).isalpha() else x) 
```
