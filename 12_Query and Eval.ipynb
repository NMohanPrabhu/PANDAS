<Font color ="blue" size="8">Query and Eval</font>

<p><b>The query() method allows you to query the DataFrame.</b>
 <p><b>   Resembles SQL query type syntax</b>
<p><b>The query() method takes a query expression as a string parameter, which has to evaluate to either True or False.</b>
<p><b>It returns the DataFrame where the result is True according to the query expression.</b>
<p><b>query() method filters rows from a DataFrame based on a specified condition.</b>

```python
import numpy as np
import pandas as pd
df=pd.read_csv("DATASETS\\odi.csv") 
df
```

```python
#Extracts desired rows out of data frame
#Selects country AUS and SA
df.query('Country in ("AUS","SA")')
```

```python
df.query("Runs >9000 & SR >90 ")      
```

```python
df.query("Runs >9000 | SR >90 ")
```

```python
df.query("Runs >9000 & SR >90 & Ave >45") 
```

```python
df1=pd.read_csv("DATASETS\\student.csv") 
df1
```

```python
#Select the rows of the sample DataFrame where (City = Patna).
df1.query('City=="Patna"')
```

```python
#Select the rows of the sample DataFrame where (GPA < 8) + (City = Patna
df1.query('GPA < 8 & City == "Patna"')
```

```python
#Select the rows of the sample DataFrame which has Dept in [ECE, CSE, IT].
df1.query("Dept in ['CSE','ECE','IT']")
```

<font color="blue" size="6" > <b>Eval:</b></font>
<p><b>You need to pass an expressing to evaluate as a string to df.eval() method
<p><b>The eval() method evaluates the string expression and returns the result.
<p><b>You can refer to specific columns by specifying the column label(s).


```python
mask = df.eval('Country in ("AUS","SA")')
df.loc[mask,:]
```

```python
# initialize list of lists
data = [['Raj',200,197,200],
        ['Sam',198,196,200], 
        ['Bala',195,194,140], 
        ['Hema',199,191,200],
       ]
# Create the pandas DataFrame
marks = pd.DataFrame(data, columns=['Name', 'Math', 'Phy', 'Chem'])
marks
```

```python
marks.eval("Cutoff = (Math/2)+(Phy/4)+(Chem/4)",inplace=True)
```
