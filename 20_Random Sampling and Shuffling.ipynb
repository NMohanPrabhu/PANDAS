<font color="green" size="4"><b>Bootstrap Sampling:</b></font>
<p>Bootstrap sampling is when you pick the same number of rows as in the dataset but with replacement.
<p>Used In:
<p>1.In machine learning algorithms such as Random Forest,Bagging
<p>2.Computing confidence intervals


```python
df = pd.read_csv('DATASETS\\hungary_chickenpox.csv')
```

```python
#frac – a float value representing the fraction of random axis items (rows or columns) to return. 
#By default, the method returns 1/N items (virtually meaning a single item), 
#where N is the overall number of items. 
#This parameter can't be used in combination with the n parameter.
df.sample(frac=1,replace=True,random_state=100)
```

```python
df.sample(frac=0.01,replace=True,random_state=100)
```

```python
df.sample(frac=0.10,replace=True,random_state=100)
```
