<font color="blue" size="6">Apply method</font>
<p>apply method allow the users to pass a function and apply it on every single value of the Pandas series.</p>


<b> Example 1 <b>

```python
import pandas as pd
import numpy as np
df = pd.read_csv('DATASETS\Grades.csv')
df
```

```python
df["Retake"] = df["Grades"].apply(lambda val: "Yes" if val< 45 else "No")
df
```

<b>Example 2</b>

```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('DATASETS\\hungary_chickenpox.csv')
```

```python
df1
```

```python
df1=df1.drop("Date",axis=1)
```

```python
df1
```

```python
df1["max_cases"]=df1.apply(np.max,axis="columns")
```

```python
df1
```

```python
median_cases=df1.apply(np.median,axis="rows")
```
