<b>Export CSV to a working directory</b>

```python
import pandas as pd
# initialize list of lists
data = [['Dhoni',200,110,74,5,55.00],
        ['Azhar',174,90,76,2,51.72], 
        ['Ganguly', 146,76,65,0,52.05], 
        ['Kohli',95,65,27,1,68.42],
        ['Dravid',79,42,33,0,53.16],
        ['Kapil', 74,39,33,0,52.70]]
# Create the pandas DataFrame
captaincy = pd.DataFrame(data, columns=['Name', 'Mat', 'Won', 'Lost', 'Tied', 'WR'])
# printdataframe.
print(captaincy)
```

<b>1.Export a Dataframe to a CSV file using df.to_csv().</b>

```python
# saving the dataframe
captaincy.to_csv('OUTPUT/file1.csv')
```

<b> 2.Saving CSV without headers and index. </b>

```python
# saving the dataframe
captaincy.to_csv('OUTPUT/file2.csv', header=False, index=False)
```

<b>3.Save the CSV file to a specified location</b>

```python
# saving the dataframe
captaincy.to_csv(r'OUTPUT\file3.csv')
```

<b> 4.Write a DataFrame to CSV file using tab separator</b>

```python
captaincy.to_csv('Users.csv', sep='\t', index=False,header=True)
```
