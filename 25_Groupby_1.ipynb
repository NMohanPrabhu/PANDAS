Pandas groupby is used for grouping the data according to the categories and applying a function to the categories.

```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('DATASETS\\titanic.csv')
```

```python
df
```

```python
# GroupbyPlass with average survival rate
df.groupby('Pclass').agg({'Survived':np.mean})
```

```python
# GroupbyPlassand sex with average survival rate and number of persons survived in each section
df.groupby(['Sex','Pclass']).agg({'Survived':[np.mean,np.sum]})
```

```python
#Sorts according to the Sex
df.groupby(['Sex','Pclass']).agg({'Survived':[np.mean,np.sum]}).reset_index()
```

```python
#(max fare-min fare) for each sex and for each pclass
df.groupby(['Sex','Pclass']).agg({'Fare':lambda x:np.max(x)-np.min(x)})
```

```python
# for each pclass and for each sex max fare min fare and difference
def mimax(x):
    return np.max(x)-np.min(x)

df.groupby(['Sex','Pclass']).agg({'Fare':[np.min,np.max,mimax]})
```
