<p><b> Pandas sort_values() can sort the data frame in Ascending or Descending order.

```python
import numpy as np
import pandas as pd
df=pd.read_csv("DATASETS\\odi.csv") 
df
```

```python
#Sorted by players who scored most 100
df.sort_values(by="100")
```

```python
df.sort_values(by="100",ascending=False)
```

```python
#Country players are sorted in decending and those who scored more 100’s are sorted in ascending
df.sort_values(by=["Country","100"],ascending=[False,True]).loc[:,["Player","Country","100"]]
```
