<font color="blue" size="8" ><b>Removing duplicates</font>

```python
import numpy as np
import pandas as pd
df=pd.read_csv("DATASETS\\Bowling.csv") 
df
```

```python
#Removes duplicates in country column field and keeps first entry
df.drop_duplicates("Country",keep="first") 
```

```python
#Keeps last entry
df.drop_duplicates("Country",keep="last")  
```

```python
#removes all duplicates
df.drop_duplicates() 
```

```python
#Removes duplicates in country and 100 column and keeps the first.
df.drop_duplicates(subset=["Country","5"],keep="first") 
```
