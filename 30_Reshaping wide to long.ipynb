<font color="blue" size="6">Reshaping wide to long</font>

<p>A dataset can be written in two different formats: wide and long. 
<p>A wide format has values that do not repeat in the first column.
<p>A long format has values that do repeat in the first column.

```python
import pandas as pd
import numpy as np
df = pd.read_csv('DATASETS\\titanic.csv')
```

```python
#wide_to_long
df = pd.melt(df, id_vars=['PassengerId','Name'], value_vars=['Survived', 'Pclass'])
```

```python
df
```

```python
#Make Long_to_wide
df_wide=pd.pivot(df,index=['PassengerId','Name'],columns=['variable'],values='value')
```

```python
df_wide
```

```python
#Convert index to columns
df_wide.reset_index()
```
