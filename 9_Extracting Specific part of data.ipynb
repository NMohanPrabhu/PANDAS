<font color="green"><b> .loc Method:</b> </font>
<p>The loc property is used to access a group of rows and columns by label(s) or a boolean array.</p>
<p>.loc[] is primarily label based, but may also be used with a boolean array.</p>
<b>Allowed inputs are:</b>
<ul style="list-style-type:disc;">
<li>A single label, e.g. 5 or 'a', (note that 5 is interpreted as a label of the index, and never as an integer position along the index).</li>
<li>A list or array of labels, e.g. ['a', 'b', 'c'].</li>
<li>A slice object with labels, e.g. 'a':'f'.</li>
<li>A Boolean array of the same length as the axis being sliced, e.g. [True, False, True].</li>
<li>A callable function with one argument (the calling Series or Data Frame) and that returns valid output for indexing (one of the above)</li></ul>



```python
#Selects and displays all rows values of column name = Player
df.loc[:,'Player']
```

```python
#Selects and displays all rows values of passed columns column Player and Country
df.loc[:,['Player','Country']]
```

```python
#Selects and displays all rows values of continuous columns from Player till Runs
df.loc[:,'Player':'Runs']
```

```python
#Selects (row index from 2 to 4) of all continuous columns from Player till Runs
df.loc[2:4,'Player':'Runs']
```

<font color="green" size="2">.iloc Method </font>
<p>.iloc[] is an indexer used for integer-location-based indexing of data in a DataFrame</p>


```python
#The row index of 0,1,2,3 are all selected and column index of 1,2,3 are all selected 
df.iloc[[0,1,2,3],[0,2,3]]
```

```python
#The contiguous row index 0 to 3 ( 3 will not be displayed) and column index 1 to 3 
#( 3 will not be selected) are selected from the df
df.iloc[0:3,1:3]
```

<font color ="green"><b>at and iat Method: </b></font>
<ul style="list-style-type:disc;">
    <li>Provides Access to single element in the database</li>
<li>Pandas at[] is used to return data in a dataframe at the passed location.</li> 
<li>The passed location is in the format [position, Column Name].</li>
<li>This method works in a similar way to Pandas loc[ ] but at[ ] is used to return an only single value and hence works faster than it.</li>
<li>Pandas iat[] method is used to return data in a dataframe at the passed location.</li> 
<li>The passed location is in the format [position in the row, position in the column].</li>
<li>This method works similarly to Pandas iloc[] but iat[] is used to return only a single value and hence works faster than it.</li>
    </ul>


```python
df.iat[1,0]
```

```python
df.at[1,'100']
```
