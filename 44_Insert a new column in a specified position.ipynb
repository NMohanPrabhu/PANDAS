<font color="blue" size="6" >Insert a new column in a specified position</font>

When a new column is created it will be placed at last. We can specify a position of index where it has to be loaded.




```python
import pandas as pd
import numpy as np
df=pd.read_csv('DATASETS/titanic.csv')
```

```python
mean_class_fare=df.groupby('Pclass')['Fare'].transform(np.mean)
```

```python
df.insert(10,"mean_class_fare",mean_class_fare)
df.head()
```
