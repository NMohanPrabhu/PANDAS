<font color="blue" size="6">Filling the Missing values</font>

In order to fill null values in a datasets, we use fillna() function these function replace NaN values with some value of their own. 

<font color="yellow" size="4" >Code #1: Filling null values with a single value </font>

```python
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample.csv")
df = pd.DataFrame(data)
df

```

```python
df.fillna(0)
```

```python
df
```

```python
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample.csv")
df = pd.DataFrame(data)
```

<font color="yellow" size="4" >Code 4: Now we are going to fill all the null values in Gender column with “No Gender”</font>

```python
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample.csv")
df = pd.DataFrame(data)
```

<font color="yellow" size="4" >Code 5:Now we are going to replace the all Nan value in the data frame with -99 value.</font> 

```python
import numpy as np
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample.csv")
df = pd.DataFrame(data)
```

```python
df.replace(to_replace = np.nan, value = -99)
```

<font color="yellow" size="4" >Code 6: Replace NaNs with median or mean of the specified columns in a given DataFrame.</font>

```python
import numpy as np
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample1.csv")
df = pd.DataFrame(data)
df
```

```python
df['Pur'].fillna(df['Pur'].median(), inplace=True)
df['Sales'].fillna(df['Sales'].mean(), inplace=True)
df
```

<font color="yellow" size="4" >Code 7:Pandas program to count the number of missing values of a specified column in a given 	DataFrame.</font>

```python
import numpy as np
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample1.csv")
df = pd.DataFrame(data)
```

```python
df['Pur'].value_counts(dropna=False).loc[np.nan]
```

<font color="yellow" size="4" >Code 8:Drop columns with  null values in a given DataFrame.</font>

```python
import numpy as np
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample1.csv")
df = pd.DataFrame(data)

df
```

```python
#To drop rows in Pur column that have null value
df[df['Pur'].notnull()] 
```

```python
#To drop rows in Pur and Sales column that have null value
df[df['Pur'].notnull() & df['Sales'].notnull()]
```
