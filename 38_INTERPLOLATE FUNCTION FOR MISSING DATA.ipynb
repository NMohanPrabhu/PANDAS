<p>Using interpolate() function to fill the missing values using linear method. 
<p>From Wikipedia, in mathematics, linear interpolation is a method of curve fitting using linear polynomials to construct new data points within the range of a discrete set of known data points.
<p>Pandas dataframe.interpolate() function is basically used to fill NA values in the dataframe or series. But, this is a very powerful function to fill the missing values. It uses various interpolation technique to fill the missing values rather than hard-coding the value.

```python
import numpy as np
import pandas as pd  
# making data frame from csv file  
data = pd.read_csv("DATASETS\\sample.csv")
df = pd.DataFrame(data)
df
```

```python
df.interpolate(method ='linear', limit_direction ='forward')
```
